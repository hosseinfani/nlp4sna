**n-gram Language Model**

```python
!pip install --upgrade nltk
import nltk
nltk.download('punkt')
from nltk.util import ngrams
from collections import Counter
from nltk.corpus import brown, movie_reviews
nltk.download('brown')
nltk.download('movie_reviews')

#print(brown.categories())
#print(len(brown.words(categories='news')))
#print(brown.sents(categories=['news'])[:5])
#print(len(brown.sents(categories=['news', 'editorial', 'reviews'])))

tokens = brown.words(categories=['news'])
vocabularly = set(tokens)
n = 1
unigrams = ngrams(tokens, n)
unigrams_freq = Counter(unigrams)
print(unigrams_freq.most_common()[:10])

n = 2
bigrams = ngrams(tokens, n)
bigrams_freq = Counter(bigrams)
print(bigrams_freq.most_common()[:10])

n = 3
trigrams = ngrams(tokens, n)
trigrams_freq = Counter(trigrams)
print(trigrams_freq.most_common()[:10])


p_mrs_given_mr_and = trigrams_freq[('Mr.', 'and', 'Mrs.')] / bigrams_freq[('Mr.', 'and')]
print(p_mrs_given_mr_and)
p_mis_given_mr_and = trigrams_freq[('Mr.', 'and', 'Mis.')] / bigrams_freq[('Mr.', 'and')]
print(p_mis_given_mr_and)
p = []
for w in vocabularly:
  #print(w)
  p.append((trigrams_freq[('Mr.', 'and', w)] / bigrams_freq[('Mr.', 'and')], w))

print(sorted(p, reverse=True)[:10])

#what word starts sentences most often?
p = []
for w in vocabularly:
  p.append((bigrams_freq[('.', w)] / unigrams_freq[('.',)], w))
print(sorted(p, reverse=True)[:10])

p = []
for w in vocabularly:
  p.append((bigrams_freq[('?', w)] / unigrams_freq[('?',)], w))
print(sorted(p, reverse=True)[:10])

p = []
for w in vocabularly:
  p_a = bigrams_freq[('.', w)] / unigrams_freq[('.',)]
  p_b = bigrams_freq[('?', w)] / unigrams_freq[('?',)]
  p.append((p_a + p_b, w))
print(sorted(p, reverse=True)[:10])

```

**Approx. Bigram Language Model**

```python
p_mr_and_mrs = trigrams_freq[('Mr.', 'and', 'Mrs.')] / len(trigrams_freq)
print(p_mr_and_mrs)

approx_bi_p_mr_and_mrs = unigrams_freq[('Mr.',)]/len(tokens) * \
              (bigrams_freq[('Mr.', 'and')]/ unigrams_freq[('Mr.',)]) * \
              (bigrams_freq[('and', 'Mrs.')]/ unigrams_freq[('and', )])
print(approx_bi_p_mr_and_mrs)

approx_uni_p_mr_and_mrs = unigrams_freq[('Mr.',)]/len(tokens) * \
              (unigrams_freq[('and',)]/len(tokens)) * \
              (unigrams_freq[('Mrs.',)]/len(tokens))
print(approx_uni_p_mr_and_mrs)
```

```python
p_mr_and_i = trigrams_freq[('Mr.', 'and', 'I')] / len(trigrams_freq)
print(p_mr_and_i)

approx_bi_p_mr_and_i = unigrams_freq[('Mr.',)]/len(tokens) * \
              (bigrams_freq[('Mr.', 'and')]/ unigrams_freq[('Mr.',)]) * \
              (bigrams_freq[('and', 'I')]/ unigrams_freq[('and', )])
print(approx_bi_p_mr_and_i)

approx_uni_p_mr_and_i = unigrams_freq[('Mr.',)]/len(tokens) * \
              (unigrams_freq[('and',)]/len(tokens)) * \
              (unigrams_freq[('I',)]/len(tokens))
print(approx_uni_p_mr_and_i)
```

```python
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE, Laplace, KneserNeyInterpolated

#We divide the corpus into 70% training and 30% testing
corpus = brown.sents(categories=['news'])
print(corpus[:10])

train = []
test = []
#can we do randomly?
for i in range(len(corpus)):
  if i % 8:
    train += [corpus[i]]
  else:
    test += [corpus[i]]
print(len(corpus))
print(len(train))
print(len(test))
print(train[:10])
print(test[:10])

#https://www.kite.com/python/docs/nltk.lm
for n in range(3, 0, -1):
  for lm in [MLE(n), Laplace(n), KneserNeyInterpolated(n)]:
    print(f'{n}-gram LM with {type(lm).__name__}:')
    try:
      e = -1
      pp = -1
      train_, vocab_ = padded_everygram_pipeline(n, train)
      lm.fit(train_, vocab_)
      #print(lm.vocab)
      #print(lm.generate(10, text_seed=['Mr.', 'and']))
      print('P(Mrs. | Mr. and) = ', lm.score('Mrs.', ['Mr.', 'and']))#yield 0 for {1,2}-gram because the model's n matters!!
      print('P(I | Mr. and) = ', lm.score('I', ['Mr.', 'and']))#yield 0 for {1,2}-gram


      n_gram_test = ngrams([t for s in test for t in s], n)
      n_grams_test_freq = Counter(n_gram_test)
      #print(n_grams_test_freq.most_common()[:10]) 
      e = lm.entropy([ng for ng in n_grams_test_freq.keys()]) 
      pp = lm.perplexity([ng for ng in n_grams_test_freq.keys()])
    #except:
    #  pass
    finally:
      print(f'entropy: {e}, perplexity: {pp}')
      print()
  
```

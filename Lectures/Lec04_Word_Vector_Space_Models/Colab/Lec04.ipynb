```
import nltk
from nltk.corpus import wordnet as wn
nltk.download('wordnet')

dog_syn = wn.synsets("dog")
print(dog_syn)

synonyms = []
antonyms = []
for syn in wn.synsets("active"):
		for l in syn.lemmas():
			synonyms.append(l.name())
			if l.antonyms():
				 antonyms.append(l.antonyms()[0].name())
print(set(synonyms))
print(set(antonyms))

vehicle = wn.synset('vehicle.n.01')
vehicle_types = list(set([w for s in vehicle.closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
print(vehicle_types)

vehicle_parts = list(set([w for s in vehicle.closure(lambda s:s.part_meronyms()) for w in s.lemma_names()]))
print(vehicle_parts)
```

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
from nltk.corpus import brown, movie_reviews
nltk.download('brown')

docs = [' '.join(t) for t in [s for s in brown.sents(categories=['news'])[:10]]]
print(docs)
vec = CountVectorizer()
X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df.transpose())
```
